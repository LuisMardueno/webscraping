```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_file = 'data_1.csv'  # Replace with your CSV file's path
df = pd.read_csv(csv_file)
df
```

```python
for index, row in df.iterrows():
    if (row['Color'] == 'Multicolor') and (row['Estampado'] == 'Floral'):
        df.at[index, 'Temporada'] = row['Temporada'].replace('no', 'Verano')
```

```python
df
```

```python
for index, row in df.iterrows():
    if (row['Longitud_manga'] == 'Manga larga') and (row['Escote'] == 'Redondo'):
        df.at[index, 'Temporada'] = row['Temporada'].replace('no', 'Invierno')
```

```python
for index, row in df.iterrows():
    if (row['Corte'] == 'Tirantes'):
        df.at[index, 'Temporada'] = row['Temporada'].replace('no', 'Verano')
```

```python
for index, row in df.iterrows():
    if (row['Longitud_manga'] == 'Túnica'):
        df.at[index, 'Temporada'] = row['Temporada'].replace('no', 'Primavera')
```

```python
import random
temporadas = ['Primavera','Verano','Otoño','Invierno']
def get_rand_temporada():
    return random.choice(temporadas)
```

```python
for index, row in df.iterrows():
    df.at[index, 'Temporada'] = row['Temporada'].replace('no', get_rand_temporada())
```

```python
df
```

```python
df = pd.DataFrame(df)

# Save the dataframe to a CSV file
df.to_csv('data_1_mod.csv', index=False)
```
