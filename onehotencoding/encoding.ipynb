```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('data1.csv')
df.head()
```

```python
df['Precio'] = df['Precio'].str.replace('$', '')
```

```python
df['Precio']
```

```python
df['Precio'] = df['Precio'].str.replace(',', '').astype('float')
```

```python
df['Precio'] = df['Precio'].astype('int64')
```

```python
df.dtypes
```

```python
df["Temporada"].unique()
```

```python

df['Material'].unique()
```

```python
df['Material'] = df['Material'].str.replace('Algodón', 'Algodon')
```

```python
df['Color'].unique()
```

```python
df['Longitud_manga'] = df['Longitud_manga'].str.replace('Manga corta', 'Corta')
df['Longitud_manga'] = df['Longitud_manga'].str.replace('Manga larga', 'larga')
df['Longitud_manga'] = df['Longitud_manga'].str.replace('Sin mangas', 'Sin_manga')
```

```python
df['Corte'] = df['Corte'].fillna('Generico')
```

```python
df['Corte'].unique()
```

```python
ohe = OneHotEncoder()
```

```python
print(ohe)
```

```python
ohe.fit_transform(df[["Temporada","Color","Estilo","Estampado","Escote","Longitud_manga","Corte","Material"]]).toarray()
```

```python
feature_array = ohe.fit_transform(df[["Temporada","Color","Estilo","Estampado","Escote","Longitud_manga","Corte","Material"]]).toarray()
```

```python
ohe.categories_
```

```python
feature_labels = ohe.categories_
print(feature_labels)
```

```python
feature_labels = [['Invierno', 'Otoño', 'Primavera', 'Verano'],
                  ['Amarillo', 'Azul', 'Blanco', 'Cafe', 'Gris', 'Morado', 'Multicolor', 'Naranja', 'Negro', 'Rojo', 'Rosa', 'Verde'],
                  ['Casual', 'Elegante', 'Fiesta', 'Sexy', 'Vintage'],
                  ['Combinado', 'Floral', 'Generico', 'Geometrico', 'Liso'],
                  ['Alzado', 'Asimetrico', 'Camisero', 'Corazon', 'Recto', 'Redondo', 'V'],
                  ['Corta', 'Media', 'Sin_manga', 'larga'],
                  ['A línea', 'Ajustado', 'Camisa', 'Generico', 'Holgado', 'Natural', 'Tirantes'],
                  ['Algodon', 'Elastano', 'Poliamida', 'Poliuretano', 'Tela tricotada', 'Viscosa']]

flattened_labels = np.array([label for sublist in feature_labels for label in sublist])

print(flattened_labels)
```

```python
np.array(flattened_labels).ravel()
```
